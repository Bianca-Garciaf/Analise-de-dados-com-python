1. Use o Pandas para acessar os arquivos `title.ratings.tsv.gz`
e `title.basics.tsv.gz`
> O que significa essa extensão dos arquivos?
uma seleciona as informações em relação aos programas e a outra está relacionada com a avaliação. Basicamente eles funcionam como uma forma de selecionar quais arquivos e tipos de arquivo voce quer extrair.

```python
ratings_url = 'https://datasets.imdbws.com/title.ratings.tsv.gz'
basics_url = 'https://datasets.imdbws.com/title.basics.tsv.gz'

# Ler os arquivos diretamente do link
db_ratings = pd.read_csv(ratings_url, sep="\t", compression="gzip")
db_basics = pd.read_csv(basics_url, sep="\t", compression="gzip")
```

```python
db_ratings
```

```python
db_basics
```

2. Quais são os principais valores presentes na coluna titleType?

```python
db_basics["titleType"].value_counts()
```

3. Filtre a tabela title basics para manter apenas entradas do tipo `tvSeries`

```python
db_basics.loc[db_basics["titleType"] == "tvSeries"]
```

4. Junte as informações das duas tabelas em uma só, com o método `.merge`

```python
db_merged = pd.merge(db_basics, db_ratings, on='tconst', how='inner')
```

```python
db_merged
```

5. Quantas séries desse dataset iniciaram em cada ano?

```python
series = db_merged[db_merged["titleType"] == "tvSeries"]
series["startYear"].value_counts()
```

6. Qual é a média de rating para as séries de cada ano?

```python
series.groupby("startYear")["averageRating"].mean()
```

7. Qual é a correlação entre o rating e (1) duração em minutos, (2) ano de início e (3) tempo que a série ficou no ar.

```python


series_corr = series.dropna(subset=["averageRating", "runtimeMinutes", "startYear", "endYear"])

series_corr["startYear"] = pd.to_numeric(series_corr["startYear"], errors="coerce").fillna(0).astype(int)
series_corr["endYear"] = pd.to_numeric(series_corr["endYear"], errors="coerce").fillna(0).astype(int)
series_corr["runtimeMinutes"] = pd.to_numeric(series_corr["runtimeMinutes"], errors="coerce").fillna(0).astype(int)


series_corr["yearsOnAir"] = series_corr["endYear"] - series_corr["startYear"]

series_corr[["averageRating", "runtimeMinutes", "startYear", "yearsOnAir"]].corr()




```
