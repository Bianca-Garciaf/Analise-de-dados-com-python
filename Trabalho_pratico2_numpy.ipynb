```python
import numpy as np
    #Foram cirados os 4 vetores com auxilio da biblioteca numpy(usada para computação númerica e operações matematicas com (arrays/vetores))

matriz = np.zeros((4,21), dtype = str)
matriz[0,0] = 'A'
matriz[1,0] = 'B'
matriz[2,0] = 'C'
matriz[3,0] = 'D'
i = 0
j = 0
contTicket = 0
contMeia = 0
contInteira = 0



print("Bem vindo ao programa de cadastro de passagens aéreas!\nInstruções:")
print(" 1-Assentos de 1 a 20 em cada fileira\n 2-Apenas números inteiros positivos\n 3-Bebés com menos de 1 ano, podem ser colocados como idade 0\n 4-O programa irá encerra com respostas diferentes de sim ou quando os 80 lugares forem ocupados\n")
comprar = str(input("Deseja cadastrar uma passagem?"))


    #Função para a confirmação da idade do passageiro e contagem do valor das passagens de acordo com a faixa étaria

def Idade(int):

  idade = int(input("Qual a idade do passageiro?"))
  global contMeia
  global contInteira

  if(idade < 5) or (idade> 65):
    print("O valor da passagem sofrera desconto de 50% (R$50)")
    contMeia = contMeia + 50

  else:
    print("Passagem inteira(R$100,00)")
    contInteira = contInteira + 100

    #Função para a escolha das fileiras e assentos, além da contagem de passagens vendidas e evitar venda da mesma poltrona
    #Usa a função Idade

def Assento(int):

  print(matriz)
  fileira= str(input("Qual fileira você deseja: A, B, C ou D? "))
  y = fileira.upper()
  global contTicket

  match y:
    case "A":
      poltrona = int(input("Digite o número da poltrona: "))
      if( poltrona>20 or matriz[0,poltrona] == '1' ):
        print("Esse assento não é válido, escolha outro")
      else:
        matriz[0,poltrona] = 1
        print(matriz)
        contTicket = contTicket +  1
        print(Idade(int))

    case "B":
      poltrona = int(input("Digite o número da poltrona: "))

      if( poltrona>20 or matriz[1,poltrona] == '1' ):
        print("Esse assento não é válido, escolha outro")
      else:
        matriz[1,poltrona] = 1
        print(matriz)
        contTicket = contTicket +  1
        print(Idade(int))

    case "C":
      poltrona = int(input("Digite o número da poltrona: "))

      if( poltrona>20 or matriz[2,poltrona] == '1' ):
        print("Esse assento não é válido, escolha outro")
      else:
        matriz[2,poltrona] = 1
        print(matriz)
        contTicket = contTicket +  1
        print(Idade(int))

    case "D":
      poltrona = int(input("Digite o número da poltrona: "))

      if( poltrona>20 or matriz[3,poltrona] == '1' ):
        print("Esse assento não é válido, escolha outro")
      else:
        matriz[3,poltrona] = 1
        print(matriz)
        contTicket = contTicket +  1
        print(Idade(int))


    #Função para exibição dos resultados sobre cancelamento da viagem, ocupação e vendas totais

def resultados(int):
    print("Foram comprada(s)", contTicket," passagem(s)")

    if(contTicket<40):
      print("Lamentamos informa que a viagem será cancelada por possuir menos de 50% da capacidade total.")

    else:
      capacidade = (contTicket*100)/80
      valorTotal = contMeia + contInteira


      print("A Viagem está confirmada!\nO avião possui ", capacidade,"% de ocupação e foi arrecadado R$",valorTotal)

    #Programa sendo executado após a primeira compra
if(contTicket<80):
  while comprar == 'Sim' or comprar == 'sim':
    print(Assento(int))
    comprar = str(input("Deseja cadastrar uma passagem?"))
  print(resultados(int))


    #Lógica após as vendas serem encerradas pelo operador ou por venda total das passagens

elif():
  print(resultados(int))










```
